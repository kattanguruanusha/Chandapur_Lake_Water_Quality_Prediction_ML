```python
import numpy as np

class FeatureEngineer:
    def build_time_features(self, df):
        df['month'] = df['date'].dt.month
        df['year'] = df['date'].dt.year
        return df

    def pollution_index(self, df):
        df['pollution_index'] = (
            df['bod'] + df['cod'] + df['turbidity'] + df['nitrate']
        ) / 4
        return df
```
