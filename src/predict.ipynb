```python
import torch

def predict_future(model, input_seq, months=12):
    model.eval()
    preds = []
    current = input_seq.clone()

    for _ in range(months):
        with torch.no_grad():
            out = model(current.unsqueeze(0))
            preds.append(out.squeeze().numpy())
            current = torch.cat([current[1:], out.unsqueeze(0)], dim=0)

    return preds
```
